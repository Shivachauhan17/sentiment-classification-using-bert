```python
import pandas as pd
```

```python
df=pd.read_csv('/content/train.csv',encoding= 'unicode_escape')
```

```python
df.head()
```

```python
mydf=df[['text','sentiment']]
```

```python
mydf.head()
```

```python
type(mydf['text'][0])
```

```python
import re
for i in range(len(mydf.text)):
  mydf['text'][i]=re.sub(r"[^a-zA-Z]"," ",str(mydf['text'][i]))
```

```python
mydf.head()
```

```python
mydf['sentiment'].value_counts()
```

```python
mydf.isnull().mean()
```

```python
label_map={"negative":0,"positive":2,"neutral":1}
```

```python
mydf['labels']=mydf['sentiment'].map(label_map)
```

```python
mydf.drop(labels=['sentiment'],axis=1,inplace=True)
```

```python
df1=mydf[mydf['labels']==1]
df2=mydf[mydf['labels']==2]
df0=mydf[mydf['labels']==0]
```

```python
#because of limited colab memory take 1500 example for each label
df1.drop(df1.index[1500:],axis=0,inplace=True)
df2.drop(df2.index[1500:],axis=0,inplace=True)
df0.drop(df0.index[1500:],axis=0,inplace=True)
```

```python
df1_train=df1.iloc[:1000]
df1_test=df1.iloc[1000:]
df2_train=df2.iloc[:1000]
df2_test=df2.iloc[1000:]
df0_train=df0.iloc[:1000]
df0_test=df0.iloc[1000:]
```

```python
df_train=pd.concat([df2_train,df0_train,df1_train])
df_test=pd.concat([df1_test,df2_test,df0_test])
```

```python
import nltk 
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('all')
```

```python
df_train=df_train.reset_index()
df_train.drop(labels=['index'],axis=1,inplace=True)
for i in range(len(df_train['text'])):
  df_train['text'][i] = [word for word in (nltk.word_tokenize(str(df_train['text'][i]))) if word not in set(stopwords.words('english'))]
    
```

```python
df_test=df_test.reset_index()
df_test.drop(labels=['index'],axis=1,inplace=True)
for i in range(len(df_test['text'])):
  df_test['text'][i] = [word for word in (nltk.word_tokenize(str(df_test['text'][i]))) if word not in set(stopwords.words('english'))]
    
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
train_args={"reprocess_input_data":True,
            "fp16":False,
            "num_train_epochs":10}
model=ClassificationModel("bert","distilbert-base-uncased",num_labels=3,args=train_args)
```

```python
df_train['labels'].unique()
```

```python
model.train_model(df_train)
```

```python
result=model.predict(str(df_test['text'][0]))
```

```python
import numpy as np 
result=np.argmax(result[0])
```
